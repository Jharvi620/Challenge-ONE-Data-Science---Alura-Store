# **Alura Store**

## **Importando datos**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url1 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_1%20.csv'
url2 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_2.csv'
url3 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_3.csv'
url4 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_4.csv'

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
```

## **Analizando Datos**

### **Calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.**

```python
IngresoTienda1 = int(tienda1['Precio'].sum())
IngresoTienda1
```

```python
IngresoTienda2 = int(tienda2['Precio'].sum())
IngresoTienda2
```

```python
IngresoTienda3 = int(tienda3['Precio'].sum())
IngresoTienda3
```

```python
IngresoTienda4 = int(tienda4['Precio'].sum())
IngresoTienda4
```

### **Calcular la cantidad de productos vendidos por categoría en cada tienda.**

```python
tienda1.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False).head()
```

```python
tienda2.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False).head()
```

```python
tienda3.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False).head()
```

```python
tienda4.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False).head()
```

### **Calcular las calificaciones promedio de los clientes para cada tienda**

```python
calificionTienda1 = tienda1['Calificación'].mean().round(2)
calificionTienda1
```

```python
calificionTienda2 = tienda2['Calificación'].mean().round(2)
calificionTienda2
```

```python
calificionTienda3 = tienda3['Calificación'].mean().round(2)
calificionTienda3
```

```python
calificionTienda4 = tienda4['Calificación'].mean().round(2)
calificionTienda4
```

### **Debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda**

```python
total = tienda1.groupby('Producto')['Producto'].count().sort_values(ascending=False)

topVentas = total.head()
lowVentas = total.tail()

resultado = pd.concat([topVentas, lowVentas], keys=['Más Vendidos', 'Menos Vendidos'])
resultado
```

```python
total = tienda2.groupby('Producto')['Producto'].count().sort_values(ascending=False)

topVentas = total.head()
lowVentas = total.tail()

resultado = pd.concat([topVentas, lowVentas], keys=['Más Vendidos', 'Menos Vendidos'])
resultado
```

```python
total = tienda3.groupby('Producto')['Producto'].count().sort_values(ascending=False)

topVentas = total.head()
lowVentas = total.tail()

resultado = pd.concat([topVentas, lowVentas], keys=['Más Vendidos', 'Menos Vendidos'])
resultado
```

```python
total = tienda4.groupby('Producto')['Producto'].count().sort_values(ascending=False)

topVentas = total.head()
lowVentas = total.tail()

resultado = pd.concat([topVentas, lowVentas], keys=['Más Vendidos', 'Menos Vendidos'])
resultado
```

### **calcular el costo de envío promedio para cada tienda**

```python
envioTienda1 = tienda1['Costo de envío'].mean().round(2)
envioTienda1
```

## **Graficos de las tiendas**

```python
#grafico de lineas de los ingreso por tienda
laibels = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
valores = [IngresoTienda1, IngresoTienda2, IngresoTienda3, IngresoTienda4]

plt.plot(laibels, valores, color='green', linestyle='-', marker='o', markersize=8, linewidth=2)

plt.title('Ingreso por tiendas', fontsize=16)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Ingresos', fontsize=12)


for i, v in enumerate(valores):
    plt.text(i, v + 0.08, str(v), ha='center', va='bottom', fontweight='bold', fontsize=10)

plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
```

```python
#grafico de columna de las calificaciones por tienda
laibels = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
valores = [calificionTienda1, calificionTienda2, calificionTienda3, calificionTienda4]
colores = ['#C8F9E8', '#FFDAB9', '#E6E6FA', '#ADD8E6']

plt.figure(figsize=(12, 6))
plt.title('Calificaciones por tiendas')
plt.xlabel('Tiendas')
plt.ylabel('Calificaciones')
plt.bar(laibels, valores, color=colores)

for i, v in enumerate(valores):
    plt.text(i, v + 0.08, str(v), ha='center', va='bottom', fontweight='bold', fontsize=10)

```

```python
#grafico de lineas de los ingreso por tienda
laibels = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
valores = [envioTienda1,envioTienda2,envioTienda3,envioTienda4]

plt.figure(figsize=(12, 6))
plt.title('Promedio de costo de envio')
plt.xlabel('Costo de envio')
plt.ylabel('Tiendas')
plt.barh(laibels, valores, color= colores )
```

## **Informe sobre el analisis**

### **Planteamiento del problema:**

El señor Juan es propietario de cuatro tiendas y ha decidido vender una de ellas. El propósito de esta venta es utilizar el capital obtenido para invertir en un nuevo negocio.

El objetivo principal es identificar cuál de las cuatro tiendas es la menos eficiente (o menos rentable), ya que esa será la seleccionada para la venta.

###

### **Resumen**
Basado en el análisis de los datos de las cuatro tiendas

*   **Ingreso Total:**
    *   Tienda 1: $1150880400
    *   Tienda 2: $1116343500
    *   Tienda 3: $1098019600
    *   Tienda 4: $1038375700

*   **Calificación Promedio del Cliente:**
    *   Tienda 1: 3.98
    *   Tienda 2: 4.04
    *   Tienda 3: 4.05
    *   Tienda 4: 4.0

*   **Costo de Envío Promedio:**
    *   Tienda 1: $26018.61
    *   Tienda 2: $25216.24
    *   Tienda 3: $24805.68
    *   Tienda 4: $23459.46


*   **Menor Ingreso Total:** La Tienda 4 tiene el ingreso total más bajo entre las cuatro tiendas.
*   **Menor Calificación Promedio del Cliente (empatada):** La Tienda 4 tiene la calificación promedio del cliente más baja, empatada con la Tienda 1. Aunque no es significativamente más baja que las otras, indica posibles áreas de mejora en la satisfacción del cliente.
*   **Menor Costo de Envío Promedio:** Si bien tener el menor costo de envío promedio podría parecer positivo, en el contexto de menores ingresos y calificación promedio, podría sugerir menos ventas o prácticas de envío menos eficientes que no se traducen en una mayor satisfacción del cliente o mayores ingresos.



### **Conclusion**

Tras analizar el rendimiento de las cuatro propiedades, la Tienda 4 se perfila como la menos eficiente y, por lo tanto, la candidata más lógica para la venta.

Esta recomendación se fundamenta en dos indicadores clave:


*   Es la que menos ingresos genera: La Tienda 4 reporta el ingreso total más bajo ($1,038,375,700), situándose notablemente por detrás de las demás.

*   Baja satisfacción del cliente: Presenta la calificación promedio de clientes más baja (4.0, empatada con la Tienda 1), lo que indica un rendimiento deficiente en el servicio o la experiencia de compra.



Aunque la Tienda 4 tiene el costo de envío promedio más bajo, este dato no representa una ventaja competitiva. Por el contrario, es probable que sea una consecuencia directa de su bajo volumen de ventas (lo que explica los bajos ingresos), y no un signo de eficiencia logística.

Vender la Tienda 4 permitiría al señor Juan desinvertir en el activo con el peor desempeño financiero y la menor aprobación de clientes, liberando así capital para su nuevo negocio.
